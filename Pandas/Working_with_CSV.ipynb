# Reading files from URL

```python
import pandas as pd
import requests

from io import StringIO 

url = "https://raw.githubusercontent.com/tiangechen/datasets/main/movies.csv"


headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)


data = StringIO(req.text) 

df = pd.read_csv(data)
df.head()
```

# Names

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",names=["id","city","date","winner"])
df.head()
```

# index_col

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",index_col="id")
df.head()
```

# Header

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",header = 1)
df.head()
```

# usecols

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",usecols=["id","city","winner"])
df.head()
```

# Skiprows and nrows

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",skiprows=[1,3,5,7,9])
df.head()
```

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",nrows=10)
df.head()
```

# Encoding

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",encoding='utf-8')
df.head()
```

# on_bad-lines

```python
df = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv",sep=" ",on_bad_lines='skip')
df.head()
```

# dtype

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",dtype={"date":'object'})
df.head()
```

# Handling Dates

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",parse_dates=['date'])
df.head()
```

# Converters

```python
def rename(name):
    if name == "Royal Challengers Banglore":
        return "RCB"
    else:
        return name

df = pd.read_csv("IPL Matches 2008-2020.csv",converters={"team1":rename})
df.head()
```

# na_values

```python
df = pd.read_csv("covid_toy.csv",na_values=["Male"])
df.head()
```

```python
df = pd.read_csv("IPL Matches 2008-2020.csv",chunksize=204)
for chunk in df:
    print(chunk.shape)
```
