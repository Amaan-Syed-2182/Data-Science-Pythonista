# Pandas

```python
import pandas as pd
s = pd.Series([54,65,55,76,87,83])
print(type(s))
print(s)

# Can Label Index of Series
s1 = pd.Series([54,65,55,76,87,83],index=["Harry","Raj","Shabaul","Saif","Rehan","Mehran"])
print(type(s1))
print(s1)
```

```python
# DataFrame using dictionaries
data = {
    "Name":['Harry',"Amaan","Shabaul","Rehan","Raj"],
    "Age":[28,23,25,18,24],
    "City":['Delhi','Udaipur','Muzaffarpur','Tonk','Udaipur']
}
df = pd.DataFrame(data)
print(type(df))
df
```

```python
# DataFrame using Array
import numpy as np
arr = np.array([[1,2],[5,6]])
df1 = pd.DataFrame(arr,columns = ['A','B'])
df1
```

```python
# DataFrame using List
l1 = [[1,2,3],[4,5,6]]
df2 = pd.DataFrame(l1,columns = ['A','B','C'])
df2
```
