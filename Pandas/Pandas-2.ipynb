```python
# USe copy() to avoid Modification reflection
# Changes int series reflected in the Dataframe too
df.head(2)
nameSeries= df["Name"].copy()
nameSeries[0] = "Jungle Fury"
print(nameSeries)
df.head(2)
```

```python
# Counting the occurence of each type
print(df["Type"].value_counts(sort = True).head(5))
print(df["Type"].value_counts(sort = True, ascending = False).head(5))
print(df["Type"].value_counts(sort = False, ascending = True).head(5)) # Sort must be true to use ascending
```

```python
print(df["Type"].value_counts(normalize = True).head(5)) # Tells the Porpotion of each type
```

```python
def func(row):# ground,grass,fire,normal
    mammal = ["Ground","Grass","Fire","Normal"]
    if(row["Type"] in mammal):
        print("Mammal")
    else:
        print("Pokemon")

category = df.apply(func,axis=1)
print(category)
```

```python
new_type = df["Type2"] = df["Type"].apply(lambda x:x.split(","))
print(new_type.head(5))
```

```python
df[["Type1","Type2"]] = df["Type"].str.split(",",expand=True)
print(df[["Type1","Type2"]].head(5))
```
