# Data Selection and Filtering

```python
import pandas as pd
df = pd.read_csv("Movie.csv")
df
```

```python
# Selectinng Columns
# One column
print(df['Actor'])
```

```python
# selecting specific column value
df['Actor'][0]
```

```python
# Selectiong Multiple
df[["Actor","Film"]]
```

# iloc() and loc()

```python
# Selecting Rows Using Index position
df.iloc[4]
```

```python
# Selecting Rows using Index Position
df.loc[0]
```

```python
# Slectiong specific row value
df.iloc[1,1]
```

```python
# selecting Specific Row Value
df.loc[1,"Film"]
```

```python
# slecting using slicing
df.loc[1:5,["Actor","Film","IMDb"]]
```

```python
df.loc[3:8,["Actor","Film",'Genre','IMDb']]
```

```python
df.iloc[3:8,0::2]
```

# Selection and Filtering

```python
# Filtering on singlr condition
df[df['IMDb']>6]
```

```python
df[df["Year"]>2017]
```

```python
# Filtering on Multiple condition
df[(df["IMDb"]>6) & (df["Year"]>2017)]
```

```python
# Using query() Function
df.query("IMDb > 6 and Year > 2017")
```
