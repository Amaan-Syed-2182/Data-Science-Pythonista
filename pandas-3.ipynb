```python
# filling the Null Values with Mean
import pandas as pd

df1 = pd.read_csv("../Datasets/nba.csv")

m = df1["Salary"].mean()
df1["Salary"].fillna(m)
```

```python
def height_convertion(h):
    try:
        feet,inch = h.split("-")
        return float(feet) + float(inch) / 12
    except:
        return None

x = df1.copy()   

x["Height"] = x["Height"].apply(height_convertion)
print(x[x["Height"]>6.4])
```
